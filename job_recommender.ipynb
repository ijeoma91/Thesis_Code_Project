```python
# CREATE TABLE candidate
create_candidate_table = """
CREATE TABLE IF NOT EXISTS candidate (
    id SERIAL PRIMARY KEY,
    uuid UUID DEFAULT gen_random_uuid(),
    email VARCHAR(255) NULL,
    profession VARCHAR(255) NULL,
    bio VARCHAR(512) NULL,
    about_me VARCHAR(1024) NULL,
    is_student BOOLEAN DEFAULT FALSE,
    skills TEXT[] DEFAULT '{}',
    education_level VARCHAR(255) NULL,
    embeddings_generated BOOLEAN DEFAULT FALSE,
    work_experience JSONB
);
"""
cursor.execute(create_candidate_table)
connection.commit()
print("Candidate table created.")
```

#### Function to extract below from job description
- experience level
- education level

```python
import re

def process_job_description(text, title):
    # Education Mapping
    education_levels = {
        "high school": ["high school"],
        "bachelor's": ["bachelor", "b.sc", "undergraduate"],
        "master's": ["master", "m.sc"],
        "phd": ["phd", "doctorate"]
    }

    # Experience Level Mapping by Years
    experience_levels_years = {
        "intern": (0, 0),
        "junior": (1, 2),
        "intermediate": (3, 5),
        "mid": (6, 8),
        "senior": (9, 12),
        "executive": (13, float('inf'))
    }

    # Experience Level Mapping by Role Titles
    experience_levels_titles = {
        "intern": ["intern"],
        "junior": ["assistant", "coordinator", "entry-level", "technician", "associate"],
        "intermediate": ["specialist", "therapist", "architect", "engineer"],
        "mid": ["manager", "lead", "consultant", "supervisor"],
        "senior": ["senior", "director", "head"],
        "executive": ["vp", "executive", "ceo", "cfo", "president", "attorney"]
    }

    # Extract years of experience
    experience_years = [int(year) for year in re.findall(r"(\d+)\s+year", text.lower())]
    max_experience = max(experience_years) if experience_years else 0

    # Determine experience level by title
    title = title.lower()
    experience_level = None
    for level, keywords in experience_levels_titles.items():
        if any(keyword in title for keyword in keywords):
            experience_level = level
            break

    # Fallback to years if no level found from title
    if not experience_level:
        experience_level = "junior"  # Default
        for level, (min_years, max_years) in experience_levels_years.items():
            if min_years <= max_experience <= max_years:
                experience_level = level
                break

    # Extract education level
    education = "bachelor's"  # Default if nothing is found
    for level, keywords in education_levels.items():
        if any(keyword in text.lower() for keyword in keywords):
            education = level
            break

    return {
        "num_years_experience": max_experience,
        "experience_level": experience_level,
        "education_level": education
    }
```

#### Function to carry out extraction on entire dataframe

```python
def add_experience_and_education_levels(df):
    # Define the function to process individual rows
    def process_row(row):
        if pd.isna(row['title']) or pd.isna(row['description']):
            return {"experience_level": np.nan, "num_years_experience": np.nan, "education_level": np.nan}

        result = process_job_description(row['description'], row['title'])
        return result

    # Apply the processing function row-wise
    results = df.apply(lambda row: process_row(row), axis=1)

    # Convert the results into a DataFrame
    processed_df = pd.DataFrame(results.tolist(), index=df.index)

    # Merge the results back into the original DataFrame
    df = pd.concat([df, processed_df], axis=1)
    return df
```

#### Try using regex to extract needed candidate details

It was observed that there was a lot of inconsistency in extracted data using regex which is why it is decided to use an LLM model


```python
def extract_candidate_details(text):
    # Define experience levels
    experience_levels_years = {
        "intern": (0, 0),
        "junior": (1, 2),
        "intermediate": (3, 5),
        "mid": (6, 8),
        "senior": (9, 12),
        "executive": (13, float('inf'))
    }

    def calculate_experience_level(years):
        for level, (min_years, max_years) in experience_levels_years.items():
            if min_years <= years <= max_years:
                return level
        return "junior"  # Default to junior if not matched

    # Extract profession (first line before "/")
    profession_match = re.match(r"^(.*?)/", text)
    profession = profession_match.group(1).strip() if profession_match else "Unknown"

    # Extract 'About Me' (first block of summary details)
    about_me_match = re.search(r"Summary\s+(.*?)Highlights", text, re.DOTALL)
    about_me = about_me_match.group(1).strip() if about_me_match else "Not Available"

    # Extract experience with responsibilities and skills from responsibilities
    experience = {}
    skills_set = set()
    experience_blocks = re.findall(
        r"(.*?)\s+(\w+\s+\d{4})\s+to\s+(\w+\s+\d{4}|Current)(.*?)(?=\n\n|\Z)", text, re.DOTALL
    )
    for role, start_date, end_date, responsibilities in experience_blocks:
        role = role.strip()
        start_date = start_date.strip()
        end_date = end_date.strip()
        responsibilities_list = [
            resp.strip() for resp in re.split(r"\.\s+", responsibilities.strip()) if resp
        ]
        # Extract potential skills from responsibilities
        for resp in responsibilities_list:
            potential_skills = re.findall(r"\b[A-Za-z][A-Za-z0-9&+\-/]+\b", resp)
            skills_set.update(map(str.lower, potential_skills))  # Add to skills (case-insensitive)

        date_range = f"{start_date} to {end_date}"
        experience[date_range] = {"role": role, "responsibilities": responsibilities_list}

    # Extract skills explicitly listed in the "Skills" section
    skills_match = re.search(r"Skills\s+(.*?)\n\n", text, re.DOTALL)
    if skills_match:
        explicit_skills = [
            skill.strip().lower() for skill in skills_match.group(1).split(",")
        ]
        skills_set.update(explicit_skills)

    # Calculate total years of experience
    total_years = 0
    for date_range in experience.keys():
        start, end = date_range.split(" to ")
        try:
            start_date = datetime.strptime(start, "%b %Y")
            end_date = datetime.now() if end == "Current" else datetime.strptime(end, "%b %Y")
            total_years += (end_date - start_date).days / 365.25
        except ValueError:
            pass  # Skip invalid date formats

    experience_level = calculate_experience_level(round(total_years))

    return {
        "profession": profession,
        "about_me": about_me,
        "skills": list(skills_set),
        "experience": experience,
        "experience_level": experience_level
    }
```

Below similarity algorithms were utilized in evluation
- cosine
- adjusted cosine
- soft cosine
- l1 distance
- l2 distance
- max inner product

#### Analysis of evaluation

- Accuracy: Measures overall correctness. Higher is better.

- - Best: Soft_cosine and Adjusted_cosine (both 0.55).

- F1-Score: Balances precision and recall. Higher is better.

- - Best: Soft_cosine and Adjusted_cosine (both ~0.182).

- Precision: Measures the correctness of positive predictions. Higher is better if false positives are more critical.

- - Best: Soft_cosine and Adjusted_cosine (both 1.0).

- Recall: Measures how well positives are identified. Higher is better if false negatives are more critical.

- - Best: All metrics are tied (0.1).


##### Conclusion:
Soft_cosine and Adjusted_cosine are the best overall.
They achieve the highest accuracy, F1-score, and precision while maintaining the same recall as others.




```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine as adjusted_cosine
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
from scipy.spatial.distance import cityblock  # L1 distance
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from scipy.spatial.distance import pdist, squareform  # For Soft Cosine
from sklearn.metrics import classification_report, roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

class CandidateMatcher:
    def __init__(self, job_embeddings, candidate_embeddings, labeled_data):
        """
        Initialize the matcher with preloaded job and candidate embeddings and labeled data.

        Args:
            job_embeddings (dict): A dictionary of job_id -> embedding.
            candidate_embeddings (dict): A dictionary of candidate_id -> embedding.
            labeled_data (list): List of dictionaries with candidate_id, job_id, and label.
        """
        self.job_embeddings = job_embeddings  # Preloaded job embeddings
        self.candidate_embeddings = candidate_embeddings  # Preloaded candidate embeddings
        self.labeled_data = labeled_data  # Labeled data for evaluation

    def compute_cosine_similarity(self, candidate_embedding, job_embeddings):
        return cosine_similarity([candidate_embedding], job_embeddings)[0]

    def compute_max_inner_product(self, candidate_embedding, job_embeddings):
        return np.dot(job_embeddings, candidate_embedding)

    def compute_l2_distance(self, candidate_embedding, job_embeddings):
        return -euclidean_distances([candidate_embedding], job_embeddings)[0]  # Negated for ranking

    def compute_l1_distance(self, candidate_embedding, job_embeddings):
        return -np.array([cityblock(candidate_embedding, job) for job in job_embeddings])  # Negated for ranking

    def compute_soft_cosine_similarity(self, candidate_embedding, job_embeddings):
        similarity_matrix = 1 - squareform(pdist(np.vstack([candidate_embedding] + job_embeddings), metric='cosine'))
        return similarity_matrix[0, 1:]  # Return similarities to all jobs

    def compute_adjusted_cosine_similarity(self, candidate_embedding, job_embeddings):
        return 1 - pairwise_distances([candidate_embedding], job_embeddings, metric=adjusted_cosine)[0]

    def get_top_matches(self, candidate_id, similarity_metric="cosine", n=10):
        """
        Get top matches for a candidate using a specified similarity metric.

        Args:
            candidate_id (int): The ID of the candidate.
            similarity_metric (str): Similarity metric to use ('cosine', 'adjusted_cosine', 'soft_cosine').
            n (int): Number of top matches to return.

        Returns:
            list: List of top matching job IDs with scores.
        """
        candidate_embedding = self.candidate_embeddings.get(candidate_id)
        if candidate_embedding is None:
            print(f"No embedding found for candidate ID: {candidate_id}")
            return []

        job_ids = list(self.job_embeddings.keys())
        job_embeddings = np.array(list(self.job_embeddings.values()))


        # Compute similarity/distance
        if similarity_metric == 'cosine':
            scores = self.compute_cosine_similarity(candidate_embedding, job_embeddings)
            # print(scores)
        elif similarity_metric == 'inner_product':
            scores = self.compute_max_inner_product(candidate_embedding, job_embeddings)
        elif similarity_metric == 'l2':
            scores = self.compute_l2_distance(candidate_embedding, job_embeddings)
        elif similarity_metric == 'l1':
            scores = self.compute_l1_distance(candidate_embedding, job_embeddings)
        elif similarity_metric == 'soft_cosine':
            scores = self.compute_soft_cosine_similarity(candidate_embedding, job_embeddings)
        elif similarity_metric == 'adjusted_cosine':
            scores = self.compute_soft_cosine_similarity(candidate_embedding, job_embeddings)
        else:
            raise ValueError(f"Unsupported metric: {similarity_metric}")

        top_indices = np.argsort(scores)[::-1][:n]
        top_matches = [{"job_id": job_ids[i], "score": scores[i]} for i in top_indices]
        return top_matches

    def evaluate(self, metric):
        true_labels = []
        predicted_labels = []
        scores = []  # For AUC and ROC calculations

        for entry in self.labeled_data:
            candidate_id = entry["candidate_uuid"]
            true_job_id = entry["job_uuid"]
            true_labels.append(entry["label"])

            # Get the top match for the candidate
            top_match = self.get_top_matches(candidate_id, similarity_metric=metric, n=1)
            if top_match:
                predicted_job_id = top_match[0]["job_id"]
                predicted_score = top_match[0]["score"]
            else:
                predicted_job_id = None
                predicted_score = 0  # Default score for no match

            # Determine the predicted label (1 if matches, 0 otherwise)
            predicted_labels.append(1 if predicted_job_id == true_job_id else 0)
            scores.append(predicted_score)

        # Metrics calculation
        metrics = {
            "accuracy": accuracy_score(true_labels, predicted_labels),
            "f1_score": f1_score(true_labels, predicted_labels),
            "precision": precision_score(true_labels, predicted_labels),
            "recall": recall_score(true_labels, predicted_labels),
        }

        # Generate classification report
        metrics["classification_report"] = classification_report(true_labels, predicted_labels)

        # Plot confusion matrix
        self.plot_confusion_matrix(true_labels, predicted_labels, metric)

        # Plot ROC and calculate AUC
        self.plot_roc_curve(true_labels, scores, metric)

        return metrics

    def plot_confusion_matrix(self, true_labels, predicted_labels, metric):
        cm = confusion_matrix(true_labels, predicted_labels)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm)
        disp.plot(cmap=plt.cm.Blues)
        plt.title(f"Confusion Matrix for {metric.capitalize()} Similarity")
        plt.show()

    def plot_roc_curve(self, true_labels, scores, metric):
        if len(set(true_labels)) < 2:  # Handle cases with no positive/negative labels
            print(f"Cannot compute ROC/AUC for {metric}, insufficient label variety.")
            return

        fpr, tpr, _ = roc_curve(true_labels, scores)
        roc_auc = auc(fpr, tpr)

        plt.figure()
        plt.plot(fpr, tpr, color="darkorange", lw=2, label=f"AUC = {roc_auc:.2f}")
        plt.plot([0, 1], [0, 1], color="navy", lw=2, linestyle="--")
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title(f"ROC Curve for {metric.capitalize()} Similarity")
        plt.legend(loc="lower right")
        plt.show()

    def evaluate_all_metrics(self):
        metrics_list = ["cosine", "inner_product", "l2", "l1", "soft_cosine", "adjusted_cosine"]
        evaluation_results = {}

        for metric in metrics_list:
            print(f"Evaluating {metric} similarity...")
            evaluation_results[metric] = self.evaluate(metric)
            print(f"Results for {metric}: {evaluation_results[metric]}")
            print("-" * 50)

        return evaluation_results
```
